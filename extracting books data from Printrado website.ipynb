```python
from urllib.request import urlopen, Request
from urllib.parse import urlparse

from bs4 import BeautifulSoup
import datetime
import random
import re

import pandas as pd
```

```python
def getLinks(link):
    html = urlopen('https://en.wikipedia.org{}'.format(link))
    bs = BeautifulSoup(html, 'html.parser')
    content = bs.find('div', id='bodyContent')
    links_list = content.find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))
    pages = set()
    for link in links_list:
        pages.add(link['href'])
    # pages = list(set(link['href'] for link in links_list if link.has_attr('href')))
    print("page:", link['title'], "==>", len(pages) , "sub pages.")

    return list(pages)
```

```python
def wikiSpeedrun(start_page, target_page):    
    
    visited = [] # List for visited nodes.
    queue = []   # Initialize a queue
    bfs_path = [start_page[6::]]     # the path
        
    def BFS(node, target): #function for BFS
      visited.append(node)
      queue.append(node)
    
      while queue:          # Creating loop to visit each node
        m = queue.pop(0) 
        bfs_path.append(m[6::])
        sub_pages = getLinks(m)
        if target in sub_pages:
            bfs_path.append(target[6::])
            return bfs_path
          
        for page in sub_pages:
          if page not in visited:
            visited.append(page)
            queue.append(page)

    return BFS(start_page, target_page)
```

```python
def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc == '':
            internalLinks.add(f'{scheme}://{netloc}/{link.attrs["href"].strip("/")}')
        elif parsed.netloc == netloc:
            internalLinks.add(link.attrs['href'])
    return list(internalLinks)
headers = {'User-Agent': 'Mozilla/5.0'}
req = Request("https://printrado.com/", headers=headers)

# Open the URL
html = urlopen(req)
bs = BeautifulSoup(html, "html.parser")

# Get internal links
links = getInternalLinks(bs, "https://printrado.com/")
bs
```
